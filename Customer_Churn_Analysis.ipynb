```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/shikh/OneDrive/Desktop/cleaned_ecommerce_customers_dataset.csv")

df.head()
```

```python
df.describe()
```

```python
## Data Overview:

#Dataset contains 50,000 customer records
#Includes demographic, behavioral, and transactional features
#Data is clean with handled missing values
```

```python
churn_rate = df['Churned'].mean() * 100
print("Churn Rate:", churn_rate)
```

```python
## Churn Rate:

#Approximately XX% of customers have churned, indicating a significant portion of users leaving the platform.
```

```python
df.groupby('Login_Category')['Churned'].mean()
```

```python
## Login Behavior Analysis:

Customers with lower login frequency show higher churn rates, indicating that engagement is a key factor in customer retention.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Login_Category', y='Churned', data=df)
plt.title("Churn Rate by Login Category")
plt.show()
```

```python
df.groupby('Recency_Category')['Churned'].mean()
```

```python
## Recency Analysis:

Customers who have not made a purchase recently show a significantly higher churn rate.

This indicates that inactivity is a strong signal of potential churn.
```

```python
sns.barplot(x='Recency_Category', y='Churned', data=df)
plt.title("Churn Rate by Recency Category")
plt.show()
```

```python
df.groupby('Total_Purchases')['Churned'].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Total_Purchases', y='Churned', data=df)
plt.title("Purchases vs Churn")
plt.show()
```

```python
## Purchase Behavior Analysis:

Customers with fewer purchases are more likely to churn, while customers with higher purchase activity tend to stay longer.

This indicates that purchase frequency is a strong indicator of customer retention.
```

```python
sns.boxplot(x='Churned', y='Average_Order_Value', data=df)
plt.show()
```

```python
## Business Recommendations:

- Improve user engagement through notifications and personalized offers
- Target inactive users with re-engagement campaigns
- Reduce cart abandonment with better checkout experience
- Offer loyalty programs for frequent buyers
```

```python
## Conclusion:

Customer churn is highly influenced by engagement and activity levels. 
By focusing on increasing user interaction and targeting inactive users, businesses can significantly improve customer retention.
```
